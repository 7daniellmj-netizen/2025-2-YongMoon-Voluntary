## 3Ô∏è‚É£ Ïã§Ïäµ 1: Î¨¥Ï°∞Í±¥ ÏôºÏ™ΩÏúºÎ°úÎßå Í∞ÄÎäî ÏóêÏù¥Ï†ÑÌä∏

Í∞ÄÏû• Îã®ÏàúÌïú Ï†ÑÎûµ: Ìï≠ÏÉÅ ÏôºÏ™Ω(0)ÏúºÎ°úÎßå Ïù¥ÎèôÌï©ÎãàÎã§.

```python
def left_only_agent(num_episodes=5):
    """
    Î¨¥Ï°∞Í±¥ ÏôºÏ™ΩÏúºÎ°úÎßå Í∞ÄÎäî ÏóêÏù¥Ï†ÑÌä∏
    """
    # ‚úèÔ∏è TODO: ÌôòÍ≤Ω ÏÉùÏÑ±
    env = gym.make('CartPole-v1')
    episode_rewards = []

    print("üî¥ Î¨¥Ï°∞Í±¥ ÏôºÏ™ΩÏúºÎ°úÎßå Í∞ÄÎäî ÏóêÏù¥Ï†ÑÌä∏")
    print("="*50)

    for episode in range(num_episodes):
        # ‚úèÔ∏è TODO: ÌôòÍ≤Ω Ï¥àÍ∏∞Ìôî
        observation, info = env.reset()

        episode_reward = 0

        for step in range(500):  # ÏµúÎåÄ 500Ïä§ÌÖù
            # ‚úèÔ∏è TODO: Ìï≠ÏÉÅ ÏôºÏ™ΩÏúºÎ°ú Ïù¥Îèô (hint: 0 - ÏôºÏ™Ω, 1 - Ïò§Î•∏Ï™Ω)
            action = 0

            # ‚úèÔ∏è TODO: ÌñâÎèô Ïã§Ìñâ Î∞è Í≤∞Í≥º Î∞òÌôò
            observation, reward, terminated, truncated, info = env.step(action)

            episode_reward += reward

            if terminated or truncated:
                break

        episode_rewards.append(episode_reward)
        print(f"Episode {episode+1}: {episode_reward:.0f}Ï†ê")

    env.close()

    avg_reward = np.mean(episode_rewards)
    print(f"\n‚≠ê ÌèâÍ∑† Ï†êÏàò: {avg_reward:.2f}Ï†ê")
    return episode_rewards, avg_reward

# Ïã§Ìñâ
left_rewards, left_avg = left_only_agent()
```

```python
def render_and_save_gif(agent_type, filename, num_steps=500):
    """
    ÏóêÏù¥Ï†ÑÌä∏ Ïã§ÌñâÏùÑ GIFÎ°ú Ï†ÄÏû•ÌïòÍ≥†, ÎßàÏßÄÎßâ ÌîÑÎ†àÏûÑÏóêÏÑúÏùò Ï†êÏàòÎ•º Î∞òÌôòÌï©ÎãàÎã§.
    """
    env = gym.make('CartPole-v1', render_mode='rgb_array')
    frames = []

    observation, info = env.reset()
    total_reward = 0

    for step in range(num_steps):
        # ÌîÑÎ†àÏûÑ Ï†ÄÏû•
        frame = env.render()
        frames.append(frame)

        # ÌñâÎèô ÏÑ†ÌÉù
        if agent_type == 'rule':
            action = 1 if observation[2] > 0 else 0
        elif agent_type == 'random':
            action = env.action_space.sample()
        elif agent_type == 'left': # Logic for 'always left' agent
            action = 0
        else: # Default to 'left' if agent_type is not recognized or not specified
            action = 0

        observation, reward, terminated, truncated, info = env.step(action)
        total_reward += reward

        if terminated or truncated:
            break

    env.close()

    # GIFÎ°ú Ï†ÄÏû•
    imageio.mimsave(filename, frames, fps=30)
    print(f"{filename} Ï†ÄÏû• ÏôÑÎ£å! (Ï†êÏàò: {int(total_reward)}Ï†ê, {len(frames)}ÌîÑÎ†àÏûÑ)")
    return total_reward # Return total_reward for consistency with other agents.

# Î¨¥Ï°∞Í±¥ ÏôºÏ™Ω ÏóêÏù¥Ï†ÑÌä∏ Ïã§Ìñâ ÏòÅÏÉÅ
left_agent_score = render_and_save_gif('left', 'cartpole_left_agent.gif', num_steps=200)
IPImage('cartpole_left_agent.gif')
```

## 4Ô∏è‚É£ Ïã§Ïäµ 2: ÎûúÎç§ ÏóêÏù¥Ï†ÑÌä∏

Îß§ Ïä§ÌÖùÎßàÎã§ Î¨¥ÏûëÏúÑÎ°ú ÏôºÏ™Ω(0) ÎòêÎäî Ïò§Î•∏Ï™Ω(1)ÏùÑ ÏÑ†ÌÉùÌï©ÎãàÎã§.

```python
def random_agent(num_episodes=5):
    """
    ÎûúÎç§ÌïòÍ≤å ÌñâÎèôÌïòÎäî ÏóêÏù¥Ï†ÑÌä∏
    """
    # ‚úèÔ∏è TODO: ÌôòÍ≤Ω ÏÉùÏÑ±
    env = gym.make('CartPole-v1')

    episode_rewards = []

    print("üü† ÎûúÎç§ ÏóêÏù¥Ï†ÑÌä∏")
    print("="*50)

    for episode in range(num_episodes):
        # ‚úèÔ∏è TODO: ÌôòÍ≤Ω Ï¥àÍ∏∞Ìôî
        observation, info = env.reset()


        episode_reward = 0

        for step in range(500):
            # ‚úèÔ∏è TODO: ÎûúÎç§ ÏÑ†ÌÉù
            action = env.action_space.sample()


            # ‚úèÔ∏è TODO: ÌñâÎèô Ïã§Ìñâ Î∞è Í≤∞Í≥º Î∞òÌôò
            observation, reward, terminated, truncated, info = env.step(action)


            episode_reward += reward

            if terminated or truncated:
                break

        episode_rewards.append(episode_reward)
        print(f"Episode {episode+1}: {episode_reward:.0f}Ï†ê")

    env.close()

    avg_reward = np.mean(episode_rewards)
    print(f"\n‚≠ê ÌèâÍ∑† Ï†êÏàò: {avg_reward:.2f}Ï†ê")
    return episode_rewards, avg_reward

# Ïã§Ìñâ
random_rewards, random_avg = random_agent()
```

```python
# ÎûúÎç§ ÏóêÏù¥Ï†ÑÌä∏ Ïã§Ìñâ ÏòÅÏÉÅ
render_and_save_gif('random', 'cartpole_random_agent.gif', num_steps=200)
IPImage('cartpole_random_agent.gif')
```

## 5Ô∏è‚É£ Ïã§Ïäµ 3: Í∑úÏπô Í∏∞Î∞ò(ifÎ¨∏) ÏóêÏù¥Ï†ÑÌä∏

**Í∞ÑÎã®Ìïú Í∑úÏπô:**
- ÎßâÎåÄÍ∞Ä Ïò§Î•∏Ï™ΩÏúºÎ°ú Í∏∞Ïö∏Î©¥(Í∞ÅÎèÑ > 0) ‚Üí Ïò§Î•∏Ï™ΩÏúºÎ°ú Ïù¥Îèô
- ÎßâÎåÄÍ∞Ä ÏôºÏ™ΩÏúºÎ°ú Í∏∞Ïö∏Î©¥(Í∞ÅÎèÑ < 0) ‚Üí ÏôºÏ™ΩÏúºÎ°ú Ïù¥Îèô

Ïù¥Î†áÍ≤å ÌïòÎ©¥ ÎßâÎåÄÍ∞Ä Í∏∞Ïö∏Ïñ¥ÏßÄÎäî Î∞©Ìñ•ÏúºÎ°ú Ïπ¥Ìä∏Î•º ÏõÄÏßÅÏó¨ÏÑú Í∑†ÌòïÏùÑ Ïú†ÏßÄÌï† Ïàò ÏûàÏäµÎãàÎã§!

```python
def rule_based_agent(num_episodes=5):
    """
    Í∑úÏπô Í∏∞Î∞ò ÏóêÏù¥Ï†ÑÌä∏: ÎßâÎåÄ Í∞ÅÎèÑÎ•º Î≥¥Í≥† ÌñâÎèô Í≤∞Ï†ï
    """
    env = gym.make('CartPole-v1')
    episode_rewards = []

    print("üü¢ Í∑úÏπô Í∏∞Î∞ò(ifÎ¨∏) ÏóêÏù¥Ï†ÑÌä∏")
    print("="*50)

    for episode in range(num_episodes):
        observation, info = env.reset()
        episode_reward = 0

        for step in range(500):
            # observation[2]Îäî ÎßâÎåÄ Í∞ÅÎèÑ
            pole_angle = observation[2]

            # Í∞ÑÎã®Ìïú Í∑úÏπô: ÎßâÎåÄÍ∞Ä Í∏∞Ïö∏Ïñ¥ÏßÑ Î∞©Ìñ•ÏúºÎ°ú Ïù¥Îèô
            # ‚úèÔ∏è TODO : Ï°∞Í±¥Ïóê Îî∞Î•∏ ÎßâÎåÄ Ïù¥Îèô if Î¨∏ ÏûëÏÑ±
            if pole_angle > 0:
              action = 1
            else:
              action = 0



            observation, reward, terminated, truncated, info = env.step(action)
            episode_reward += reward

            if terminated or truncated:
                break

        episode_rewards.append(episode_reward)
        print(f"Episode {episode+1}: {episode_reward:.0f}Ï†ê")

    env.close()

    avg_reward = np.mean(episode_rewards)
    print(f"\n‚≠ê ÌèâÍ∑† Ï†êÏàò: {avg_reward:.2f}Ï†ê")
    return episode_rewards, avg_reward

# Ïã§Ìñâ
rule_rewards, rule_avg = rule_based_agent()
```
